```python
import numpy as np
import matplotlib.pyplot as plt
from mpl_toolkits import mplot3d
from numba import jit
```

```python
@jit(nopython=True)
def random_walk(n):
    poslist = np.zeros(n+1)
    pos = 0
    movement = np.random.randint(-1,2,size=n)
    for i in range(n):
        pos += movement[i]
        poslist[i+1] = pos
    return poslist
    
    
```

```python
N = 1000000
R = random_walk(N)

plt.figure(figsize = (20,15))
plt.plot(R)
plt.scatter(0,0,s = 200,color = 'k')
plt.scatter(N,R[-1],s = 200,color = 'r')
plt.show()
```

```python
@jit(nopython=True)
def two_d_random_walk(n):
    xposlist = np.zeros(n+1)
    yposlist = np.zeros(n+1)
    xpos = 0
    ypos = 0
    xmovement = np.random.randint(-1,2,size=n)
    ymovement = np.random.randint(-1,2,size=n)
    for i in range(n):
        xpos += xmovement[i]
        ypos += ymovement[i]
        xposlist[i+1] = xpos
        yposlist[i+1] = ypos
    return xposlist,yposlist
```

```python
N2 = 1000000
R2 = two_d_random_walk(N2)
X = R2[0]
Y = R2[1]

plt.figure(figsize = (20,15))
plt.plot(X,Y,alpha=0.5)
plt.scatter(X,Y,alpha=0.25,s=0.05);
plt.scatter(0,0,s = 200,color = 'k')
plt.scatter(X[-1],Y[-1],s = 200,color = 'r')
plt.show()
```

```python
@jit(nopython=True)
def three_d_random_walk(n):
    xposlist = np.zeros(n+1)
    yposlist = np.zeros(n+1)
    zposlist = np.zeros(n+1)
    xpos = 0
    ypos = 0
    zpos = 0
    xmovement = np.random.randint(-1,2,size=n)
    ymovement = np.random.randint(-1,2,size=n)
    zmovement = np.random.randint(-1,2,size=n)
    for i in range(n):
        xpos += xmovement[i]
        ypos += ymovement[i]
        zpos += zmovement[i]
        xposlist[i+1] = xpos
        yposlist[i+1] = ypos
        zposlist[i+1] = zpos
    return xposlist,yposlist,zposlist
```

```python
fig = plt.figure(figsize=(15,15))
ax = plt.axes(projection='3d')

N3 = 1000000
R3 = three_d_random_walk(N3)
XX = R3[0]
YY = R3[1]
ZZ = R3[2]

ax.plot(XX,YY,ZZ,alpha=0.25)
ax.scatter(XX,YY,ZZ,alpha=0.125,s=0.05);
ax.scatter(0,0,0,color = 'k',s=100)
ax.scatter(XX[-1],YY[-1],ZZ[-1],color = 'r',s=100)
```
